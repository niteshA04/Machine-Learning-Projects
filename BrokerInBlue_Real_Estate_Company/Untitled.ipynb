```python
def is_float(x):
    try:
        float(x)
    except:
        return False
    return True
```

```python
df3[~df3['total_sqft'].apply(is_float)].sample(10)
```

```python
def convert_sqft_to_num(x):
    tokens = x.split('-')
    if len(tokens) == 2:
        return(float(tokens[0])+float(tokens[1]))/2
    try:
        return float(x)
    except:
        return None
```

```python
convert_sqft_to_num('2516')
```

```python
convert_sqft_to_num('2516 - 5612')
```

```python
convert_sqft_to_num('1574Sq. Yards')
```

```python
df4 = df3.copy()
df4['total_sqft'] = df4['total_sqft'].apply(convert_sqft_to_num)
df4.sample(10)
```

### Feature Engineering

```python
df5 = df4.copy()
df5['price_per_sqft'] = df5['price']*10000/df5['total_sqft']
df5.sample(10)
```

```python
df5.isnull().sum()
```

```python
df5.shape
```

```python
len(df5.location.unique())
```

```python
df5.location = df5.location.apply(lambda x: x.strip())
location_stats = df5.groupby('location')['location'].agg('count').sort_values(ascending=False)
location_stats
```

```python
len(location_stats[location_stats<=10])
```

```python
location_less_10 = location_stats[location_stats<=10]
location_less_10
```

```python
df5.location = df5.location.apply(lambda x: 'other' if x in location_less_10 else x)
```

```python
len(df5.location.unique())
```

```python
df5 = df5.drop(['size'], axis='columns')
```
